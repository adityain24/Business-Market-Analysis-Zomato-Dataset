# Cuisine Performance Analysis
# Identify which cuisines receive the highest aggregate ratings across countries and cities, and determine whether popularity (Votes) aligns with customer satisfaction.

```python
final_df.columns
```

```python
## In need of 3 columns Highest Aggregate Rating, Country, City and Votes.
```

```python
cuisine_rate = final_df.groupby(['Cuisines'],as_index=False).agg(avg_rating=('Aggregate rating', 'mean'),
                                                  total_votes=('Votes', 'sum'),
                                                  restaurant_count=('Restaurant ID', 'count')).sort_values('avg_rating', ascending=False)
```

```python
cuisine_rate.head(10)
```

## Locality-Level Rating Distribution
## Analyze rating patterns across localities to identify high-value neighborhoods for restaurant businesses.

```python
final_df.columns
```

```python
rating_benchmark=final_df.groupby(['City'],as_index=False).agg(avg_rating=('Aggregate rating','mean'),total_votes=('Votes','sum')).sort_values('avg_rating', ascending=False)
rating_benchmark['avg_rating'] = rating_benchmark['avg_rating'].round(2)
```

```python
rating_benchmark
```

## Country-Wise Restaurant Ecosystem Comparison
## Analyze how restaurant density, average ratings, and price ranges differ by country to understand regional dining trends.

```python
final_df.columns
```

```python
eco=final_df.groupby(['Country'],as_index = False).agg(avg_rating=('Aggregate rating', 'mean'),
                                                       number_of_restaurants=('Restaurant ID','count'),
                                                       avg_cost=('Average Cost for two','mean')).sort_values('avg_rating',ascending=False)
```

```python
eco
```

## Cost vs Quality Trade-off Analysis
## Examine the relationship between Average Cost for Two and Aggregate Rating to determine whether higher prices correlate with better customer experiences.

```python
final_df.columns
```

```python
cost_df = final_df[final_df['Aggregate rating'] > 0][['Average Cost for two', 'Aggregate rating']]
```

```python
cost_df
```

```python
cost_df = final_df[final_df['Aggregate rating'] > 0][
    ['Average Cost for two', 'Aggregate rating']
]

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(
    cost_df['Average Cost for two'],
    cost_df['Aggregate rating']
)
plt.xlabel('Average Cost for Two')
plt.ylabel('Aggregate Rating')
plt.title('Cost vs Aggregate Rating (Costâ€“Quality Trend)')
plt.show()
```
